```python
import pandas as pd
a=pd.read_excel("MIDMARKS-MINOR1-EXAM.xlsx")
a
```

```python
a['SECTION'] = a['SECTION'].fillna('ZETA') 
a['SECTION'] = a['SECTION'].replace('', 'ZETA')
a
```

```python
roman_replacements = {
    "I": "1",
    "II": "2",
    "III": "3",
    "IV": "4",
    "V": "5"
}

a.replace(roman_replacements, inplace=True)
a
```

```python
columns_to_fix = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]
a[columns_to_fix] = a[columns_to_fix].replace({"AB": "o", "MP": "o"})
```

```python
row_averages = a[['DV', 'M-II', 'PP', 'BEEE', 'FL', 'FIMS']].mean(axis=1)
    
# Fill missing values with the row average
for col in ['DV', 'M-II', 'PP', 'BEEE', 'FL', 'FIMS']:
    a[col] = a[col].fillna(row_averages)
    
```

# Filled with sigma section

```python
import pandas as pd

# Load the Excel file
file_path = 'MIDMARKS-MINOR1-EXAM.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Specify the columns to sum up
columns_to_sum = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]

# Convert columns to numeric, coercing errors to NaN
df[columns_to_sum] = df[columns_to_sum].apply(pd.to_numeric, errors='coerce')

# Create a new column "Total Value" by summing up the specified columns
df["Total Value"] = df[columns_to_sum].sum(axis=1)

# Save the updated DataFrame back to an Excel file
output_file_path = 'MIDMARKS_WITH_TOTAL.xlsx'  # Replace with your desired output path
df.to_excel(output_file_path, index=False)

print(f"File with 'Total Value' column saved to {output_file_path}")
```

```python
df
```

# calculated Total Value of all subjects marks

```python
def assign_grade(total):
    if total >= 90:
        return 'A+'
    elif total >= 80:
        return 'A'
    elif total >= 70:
        return 'B+'
    elif total >= 60:
        return 'B'
    elif total >= 50:
        return 'C'
    else:
        return 'F'


df["Grade"] = df["Total Value"].apply(assign_grade)

output_file_path = 'MIDMARKS-MINOR1-EXAM.xlsx' 
df.to_excel(output_file_path, index=False)

df
```

# Grades are assigned

```python
max_total_marks = 600  
df["Percentage"] = (df["Total Value"] / max_total_marks) * 100
df["Percentage"] = df["Percentage"].round()
df.to_excel(output_file_path, index=False)

df
```
